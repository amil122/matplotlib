```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.style.use('fivethirtyeight')
```

```python
x = np.random.randint(1,15,10)
```

```python
y= np.random.randint(1,15,10)
```

```python
x
```

```python
y
```

```python
colors = [1,2,3,4,5,1,1,2,3,3]
sizes = np.arange(100,300,20)
sizes
```

```python
#we can also give the sizes individually 
#s is for determing the size of plots and c for change the colour of mark
scatter = plt.scatter(x,y , s=sizes, c= colors, edgecolor = 'k', cmap= 'pink', linewidth = 2, alpha =0.65,)
plt.grid(True)
cbar = plt.colorbar(scatter)
cbar.set_label('Satisfaction')

plt.show()
```

```python
df = pd.read_csv('data_sets/data_scatter.csv')
```

```python
df
```

```python
df['likes'].nlargest(10)
```

```python
likes = df['likes'].values
```

```python
views_count = df['view_count'].values
```

```python
ratio = df['ratio'].values
ratio
```

```python
scatter = plt.scatter(views_count, likes,  c=ratio, cmap='summer', edgecolor='k',  linewidth = 2, alpha = 0.46)
cbar = plt.colorbar(scatter)
cbar.set_label('Like/Dislike ratio')
plt.xlabel('view_count')
plt.ylabel('total like')
plt.xscale('log')
plt.yscale('log')
plt.tight_layout()
plt.show()
```
