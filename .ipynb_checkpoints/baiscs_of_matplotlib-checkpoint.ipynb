```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
ages_x = np.random.randint(22,33,10)
ages_x
```

```python
salary_y = np.linspace(40000,67000, 10, dtype = int)
```

```python
salary_py = np.linspace(34000, 70000, 10, dtype=int)
```

```python
salary_js = np.linspace(54333,82342, 10, dtype = int)
salary_js.shape
```

```python

#this is one method to add legend
# plt.legend(['all_devs','for_python'])
#'[marker][line][color]'
#one method for add the color and all #another method to add legends by using label
plt.plot(ages_x,salary_y, color = 'k', marker = 'd',linestyle = ':', label='all_devs')
#another method for adding the colour and changing the  linestyle and all 
plt.plot(ages_x,salary_py, 'b>-.' , linewidth = 4 , label='for_python')
plt.title('Salary in USD')
plt.xlabel('age')
plt.ylabel('salary')
plt.plot(ages_x, salary_js, color = 'y', marker = '*' , linestyle = '--', linewidth =3 ,label = 'for js')
#this is one method to add legend
# plt.legend(['all_devs','for_python'])
plt.grid(True)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Example data
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
sales_A = [250, 300, 400, 350, 500, 450]
sales_B = [200, 280, 340, 300, 420, 380]

# Creating a line plot
plt.plot(months, sales_A, label='Product A', marker='o',linestyle = ':')
plt.plot(months, sales_B, label='Product B', marker='x')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Monthly Sales Data')
plt.legend()
plt.show()
```

```python
df = pd.read_csv('data_sets/data_age.csv')
```

```python
df
```

```python
ages =df['Age']
```

```python
salary_all = df['All_Devs']
```

```python
salary_py = df['Python']
```

```python
salary_js = df['JavaScript']
```

```python
plt.title('Details of dev salaries')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.plot(ages,salary_all, '*-k',linewidth = 2,label ='for all dev')
plt.plot(ages, salary_py, '>-.b', label= 'for python')
plt.legend()
plt.show()
```

```python
# columns = ['All_Devs','Python','JavaScript']
# all_values = pd.concat(df[col] for col in columns)
# overall_median = all_values.median()
overall_median = 57287
```

```python
plt.plot(ages,salary_all, '*-k',linewidth = 1,label ='for all dev')
plt.fill_between(ages,salary_py,overall_median,alpha=0.2,color='darkblue', label = 'for python')
plt.legend()
```

```python
plt.plot(ages,salary_all, '*-k',linewidth = 1,label ='for all dev')
plt.fill_between(ages,salary_py,overall_median, where =(salary_py > overall_median),interpolate = True, alpha=0.2, color='darkblue', label = 'for python')
plt.legend()
```

```python
# here we're using where cluase and all for give the condition so according to that condition it will fill 
plt.plot(ages,salary_all, '*-k',linewidth = 1,label ='for all dev')
plt.fill_between(ages,salary_py,overall_median, where =(salary_py > overall_median),interpolate = True, alpha=0.2, color='darkblue', label = 'for python')
plt.legend()
```

```python
plt.plot(ages,salary_py, 'd--y', label = 'python')
plt.fill_between(ages,salary_py,overall_median, where =(salary_py > overall_median),interpolate = True, alpha=0.2, color='darkblue', label = 'for python')

plt.plot(ages,salary_py, 'd--y', label = 'python')
plt.fill_between(ages,salary_py,overall_median, where = (salary_py <= overall_median),alpha = 0.25,color='r' )
plt.legend()
```

```python
plt.plot(ages,salary_all, '*-k',linewidth = 2,label ='for all dev')
plt.plot(ages, salary_py, '>-.b', label= 'for python')
plt.fill_between(ages,salary_py, salary_all,  where = (salary_py > salary_all),alpha=0.025, interpolate = True, color ='b')
plt.fill_between(ages,salary_py,salary_all, where = (salary_py <= salary_all),color='k',)

plt.legend()
```

```python
plt.plot(ages,salary_py)
plt.plot(ages,salary_all)
plt.fill_between(ages, salary_py, salary_all, where =(salary_py > salary_all), interpolate =True)
plt.fill_between(ages, salary_py, salary_all, where =(salary_py <= salary_all), interpolate =True)
```
