```python
import matplotlib.pyplot as plt

data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5, 5, 6]
plt.hist(data, bins=6, color='skyblue', edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Basic Histogram')
plt.show()
```

```python
data1 = [1, 2, 2, 3, 3, 3, 4, 4]
data2 = [2, 3, 4, 4, 4, 5, 5, 6]
plt.hist([data1, data2], bins=6, color=['blue', 'green'], edgecolor='black', label=['Data1', 'Data2'], stacked=True)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Stacked Histogram')
plt.legend()
plt.show()
```

```python
 plt.hist(data, bins=6, color='purple', edgecolor='black', cumulative=True)
plt.xlabel('Value')
plt.ylabel('Cumulative Frequency')
plt.title('Cumulative Histogram')
plt.show()
```
