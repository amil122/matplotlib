```python
import  pandas as pd 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data_sets/data_age.csv')
df
```

```python
ages = df['Age'].values
```

```python
salary_all = df['All_Devs'].values
```

```python
salary_py = df['Python'].values
```

```python
salary_js = df['JavaScript'].values
```

```python
plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('Medain salary')
plt.plot(ages,salary_all, 'd-k', label =' for all')
plt.plot(ages,salary_py, '*:y', label ='for python')
plt.plot(ages, salary_js, '<--b', label='for JavaScript')
plt.legend()
plt.show()
```

```python
fig, ((ax1,ax2),(ax3,ax4),(ax5,ax6)) = plt.subplots(nrows=3,ncols =2)
ax1.plot(ages,salary_all, 'd-k', label =' for all')
ax2.plot(ages,salary_all, 'd-k', label =' for all')
ax3.plot(ages,salary_py, '*:y', label ='for python')
ax4.plot(ages, salary_js, '<--b', label='for JavaScript')
ax5.plot(ages,salary_py, '*:y', label ='for python')
ax6.plot(ages, salary_js, '<--b', label='for JavaScript')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

ages = [22, 25, 28, 30, 35, 40, 45, 50, 55, 60]
salary_all = [35000, 42000, 48000, 53000, 58000, 62000, 67000, 71000, 73000, 75000]
salary_py = [36000, 43000, 49000, 54000, 59000, 63000, 68000, 72000, 74000, 76000]
salary_js = [37000, 44000, 50000, 55000, 60000, 64000, 69000, 73000, 75000, 77000]

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, sharex= True)

ax1.plot(ages, salary_all, 'd-k', label='for all')
ax2.plot(ages, salary_py, '*:y', label='for Python')
ax3.plot(ages, salary_js, '<--b', label='for JavaScript')

# Adding legends to each subplot
ax1.legend()
ax2.legend()
ax3.legend()

# Adding labels and titles to each subplot
ax1.set_xlabel('Age')
ax1.set_ylabel('Salary')
ax1.set_title('All Devs')

ax2.set_xlabel('Age')
ax2.set_ylabel('Salary')
ax2.set_title('Python Devs')

ax3.set_xlabel('Age')
ax3.set_ylabel('Salary')
ax3.set_title('JavaScript Devs')

# Adjust layout
plt.tight_layout()
plt.show()
```
