```python
labels = [
 'Swift',
 'SQL',
 'PHP',
 'JavaScript',
 'Java',
 'HTML/CSS',]
```

```python
slices = [
 190120,
 92108,
 236804,
 143640,
 221796,
 127920]
explode = [0,0,0.1,0.1,0.01,0.1]
```

```python
#'%.1f%%': Displays one decimal place for percentages (e.g., 25.0%).
#'%.2f%%': Displays two decimal places for percentages (e.g., 25.00%).
#'%d%%': Displays percentages as integers (e.g., 25%).
plt.tight_layout()
plt.title('Pie charts')
plt.pie(slices, labels = labels, wedgeprops = {'edgecolor':'black','linewidth':2}, explode = explode,shadow = True, autopct='%.d%%', startangle = 90)
```
