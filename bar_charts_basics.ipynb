```python
df = pd.read_csv('data_sets/data.csv')
```

```python
df
```

```python
from collections import Counter
# it's one method used to count the each values 

# Assuming df is your DataFrame and 'LanguagesWorkedWith' is a column containing strings of languages separated by ';'
language_counter = Counter(df['LanguagesWorkedWith'].apply(lambda x: x.split(';')).explode())

# This line:
# 1. Splits each string in the 'LanguagesWorkedWith' column by ';' using the split() method.
# 2. Applies the split operation to each element of the column using the apply() method.
# 3. Uses the explode() method to convert lists of languages into separate rows.
# 4. Counts the occurrences of each language using Counter.

print(language_counter)
```

```python
# another method to count the values 
ids = df['Responder_id']
lang_response = df['LanguagesWorkedWith']
```

```python
language_counter = Counter()
```

```python
for item in lang_response:
    language_counter.update(item.split(';'))

# Extract languages and their counts from language_counter
languages = []
popularity = []
for language, count in language_counter.items():
    languages.append(language)
    popularity.append(count)
print("Languages:", languages)
print("Popularity:", popularity)
```

```python
plt.title('Most popular langauge')
plt.xlabel('number of people who use')
plt.style.use('dark_background')
plt.barh(languages,popularity)
```

```python
# we're reversing it for ascending order
languages.reverse()
popularity.reverse()
plt.title('Most popular langauge')
plt.xlabel('number of people who use')
plt.style.use('dark_background')
plt.barh(languages,popularity)
```
