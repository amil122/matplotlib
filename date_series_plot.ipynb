```python
import matplotlib.pyplot as plt
from datetime import datetime,timedelta
from matplotlib import dates as mpl_dates
import pandas as pd
```

```python
plt.style.use('classic')
```

```python
dates = [
    datetime(2019,5,24),
    datetime(2019,5,25),
    datetime(2019,5,26),
    datetime(2019,5,27),
    datetime(2019,5,28),
    datetime(2019,5,29)
]
```

```python
y = [0,1,3,2,5,6]
```

```python
#gcf = get current figure it used get the current figure and in that we're using auto format x-axis date
plt.plot_date(dates,y, linestyle ='solid')
plt.grid(True)
plt.gcf().autofmt_xdate()
date_format = mpl_dates.DateFormatter(' %d,%b,%Y')
plt.gca().xaxis.set_major_formatter(date_format)
plt.show()
```

```python
df = pd.read_csv('data_sets/data_date.csv')
```

```python
df
```

```python
new_row = pd.DataFrame([{'Date':'2019-4-12','Open':8779.00,'Close':8785.12}])
```

```python
df = pd.concat([df,new_row], ignore_index = True)
```

```python
df
```

```python

df.drop(index = [14,15], inplace = True)
```

```python
df['Date']= pd.to_datetime(df['Date'])
df.sort_values('Date', inplace = True)
```

```python
price_date = df['Date'].values
price_date
```

```python
price_close = df['Close'].values
```

```python
plt.plot_date(price_date,price_close,linestyle ='solid')
plt.xlabel('Date')
plt.ylabel('Close price')
date_format = mpl_dates.DateFormatter('%d,%b,%Y')

plt.gca().xaxis.set_major_formatter(date_format)
plt.gcf().autofmt_xdate()
plt.tight_layout()
plt.show()
```
