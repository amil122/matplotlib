```python
import numpy as np
import matplotlib.pyplot as plt

# Example data
# Suppose we have data for 3 categories (e.g., 'A', 'B', 'C') over 5 time periods
time = np.arange(1, 6)  # Time periods
data = np.array([
    [3, 4, 5, 6, 7],  # Category A
    [1, 2, 3, 4, 5],  # Category B
    [2, 3, 4, 5, 6]   # Category C
])

# Stack plot
labels = ['Category A', 'Category B', 'Category C']
colors = ['red', 'green', 'blue']

plt.stackplot(time, data, labels=labels, colors=colors)
plt.xlabel('Time')
plt.ylabel('Values')
plt.title('Stack Plot Example')
plt.legend(loc='upper left')
plt.show()
```
